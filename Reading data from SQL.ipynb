```python
import sqlite3 as sq3
import pandas.io.sql as pds
import pandas as pd
```

```python
path = 'data/classic_rock.db'
con = sq3.Connection(path)
```

```python
query = '''
SELECT * FROM rock_songs
'''
observation = pd.read_sql(query, con)
observation.head()
```

```python
query = '''
SELECT Artist, Release_Year, COUNT(*) as num_songs, AVG(PlayCount) AS avg_plays
FROM rock_songs
GROUP BY Artist, Release_Year
ORDER BY num_songs desc
'''

observations = pd.read_sql(query, con)
observations.head()
```

```python
query = '''
SELECT Artist, Release_Year, COUNT(*) as num_songs, AVG(PlayCount) AS avg_plays
FROM rock_songs
GROUP BY Artist, Release_Year
ORDER BY num_songs desc
'''

observations_generator = pd.read_sql(query, 
                                     con, 
                                     coerce_float=True,
                                     parse_dates=['Release_Year'],
                                     chunksize=5
                                    )

for index, observation in enumerate(observations_generator):
    if index < 5:
        print(f'Observation index {index}'.format(index))
        display(observations)
```
